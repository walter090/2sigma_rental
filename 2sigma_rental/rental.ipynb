```python
import pandas as pd
import numpy as np
import seaborn as sns
import pprint as pp
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
training = pd.read_json('data/train.json')
print(training.shape)
training.head()
```

```python
training.isnull().sum()
```

```python
cols = list(training)
pp.pprint(cols)
```

```python
print(training['interest_level'].unique())
plt.tick_params(colors='red')
plt.hist(training['bathrooms'])
```

```python
sns.set_style('dark')
sns.countplot(x='interest_level', data=training, order=['low', 'medium', 'high'])
```

```python
training_num = training.drop(['photos', 'listing_id'], axis=1)
sns.pairplot(data=training_num)
```

```python
figure = plt.figure(figsize=(15, 15))
sns.countplot(x='bathrooms', data=training, ax=plt.subplot(211))
sns.countplot(x='bedrooms', data=training, ax=plt.subplot(212))
```

```python
training['interest_num'] = \
    np.where(training['interest_level'] == 'low', 0,
             np.where(training['interest_level'] == 'medium', 0.5, 1))
```

```python
plt.figure(figsize=(15, 15))
sns.pointplot(x='bathrooms', y='interest_num', data=training, ax=plt.subplot(211))
sns.pointplot(x='bedrooms', y='interest_num', data=training, ax=plt.subplot(212))
```
