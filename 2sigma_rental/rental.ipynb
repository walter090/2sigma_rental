```python
def clean_tokenize(description):
    import bs4 as bs
    from nltk import RegexpTokenizer
    soup = bs.BeautifulSoup(description, 'lxml')
    for tag in soup.find_all('br'):
        tag.insert(0, ' ')
    text = soup.get_text()
    tokenizer = RegexpTokenizer(r'\w+')
    return tokenizer.tokenize(text)

descriptions = training['description']
training['features_clean'] = [clean_tokenize(des) for des in descriptions]
```
